```python
df = pd.read_csv("books_data.csv")
df.head()
```

```python
df["Price"] = df["Price"].str.replace("£", "").astype(float)
df.head()
```

```python
print("Average Price:", df["Price"].mean())
print("Max Price:", df["Price"].max())
print("Min Price:", df["Price"].min())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price"], bins=10, kde=True, color="skyblue")
plt.title("Book Price Distribution")
plt.xlabel("Price (£)")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(df["Price"])
plt.title("Boxplot of Book Prices")
plt.show()
```

## EDA Summary

- Most books are priced between **£30 and £60**
- The **average price** is approximately **£45**
- There are **a few outliers** with prices above **£80**
- The dataset contains **no missing values**
- Data was clean and ready for analysis
